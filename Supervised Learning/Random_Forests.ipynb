```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df=load_iris()
data=pd.DataFrame(df.data,columns=df.feature_names)
data
```

```python
target=df.target
target=pd.DataFrame(target,columns=['target'])
target.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,target,test_size=0.2)
```

```python
rc=RandomForestClassifier(n_estimators=180,criterion='gini',max_depth=3,oob_score=True,random_state=42,n_jobs=-1)
rc.fit(x_train,y_train)
```

```python
rc.predict(x_test)
```

```python
rc.score(x_test,y_test)
```

```python
rc.feature_importances_
```

```python
rc.decision_path(x_test)
```

```python
print(classification_report(y_test,rc.predict(x_test)))
```
